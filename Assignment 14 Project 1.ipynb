```python

users = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
sessions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
products = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
transactions = pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
print(users.head())
print(sessions.head())
print(products.head())
```

```python
#12 Join users to transactions, keeping all rows from transactions and only matching rows from users (left join)
result = pd.merge(transactions,users,how='left')
print(result)
```

```python
#13 Which transactions have a UserID not in users?
result.loc[result.User.isnull(),:]
```

```python
#14 Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)
result2 = pd.merge(left=users,right=transactions,how='inner',on='UserID')
result2
```

```python
#15 Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)
result3 = pd.merge(left=users, right=transactions,how='outer')
result3
```

```python
#16 Determine which sessions occurred on the same day each user registered
from pandasql import sqldf
pysqldf = lambda q: sqldf(q, globals())

result4 = pd.merge(left=users, right=sessions,how='inner', on='UserID')

q = """SELECT *
    FROM result4
    WHERE Registered == SessionDate;
"""

print(pysqldf(q))
```

```python
#17 Build a dataset with every possible (UserID, ProductID) pair (cross join)
#print(products.head())
#print(users.head())
q = """ SELECT UserID, ProductID
        FROM users 
        CROSS JOIN products;
"""
print(pysqldf(q))
result17 =pd.DataFrame(pysqldf(q))
```

```python
#18 Determine how much quantity of each product was purchased by each user
#print(products.head())
#print(transactions.head())
#print(result17.head())

transactions.fillna(-999999, inplace=True)
transactions.UserID = transactions.UserID.astype(int)

selection = pd.merge(result17,transactions,how='left',on=['UserID','ProductID'])
#print(selection.head())
result5 = selection.groupby(['UserID','ProductID'])['Quantity'].sum()
result5.reset_index().fillna(0)
```

```python
#19
q= """SELECT t1.TransactionID AS "TransactionID_x", t1.TransactionDate AS "TransactionID_x", t1.UserID, t1.ProductID AS \
            "ProductID_x", t1.Quantity AS "Quantity_x", t2.TransactionID AS "TransactionID_y", t2.TransactionDate AS \
            "TransactionDate_y", t2.ProductID AS "ProductID_y", t2.Quantity AS "Quantity_y"
    FROM transactions t1
    JOIN transactions t2
    ON t1.TransactionID < t2.TransactionID
    

"""
print(pysqldf(q))
```

```python
#20 Join each user to his/her first occuring transaction in the transactions table
UsersTrans = pd.merge(users,transactions,how='left',on='UserID')
dfNew = UsersTrans.drop_duplicates(subset='UserID')
dfNew = dfNew.reset_index(drop=True)
dfNew
```

```python
#21. Test to see if we can drop columns
my_columns = list(dfNew.columns) 
print(my_columns)  #current columns of dfNew

#set threshold to drop NAs
print(list(dfNew.dropna(thresh=int(dfNew.shape[0] * .9), axis=1).columns)) #which columns have missing values

#which columns have any missing data
missing_info= list(dfNew.columns[dfNew.isnull().any()]) 
print(missing_info)

#count of missing data
for col in missing_info:
    num_missing = dfNew[dfNew[col].isnull() == True].shape[0] 
    print('number missing for column {}: {}'.format(col, num_missing))
    
#percentage of missing data
for col in missing_info:
    percent_missing = dfNew[dfNew[col].isnull() == True].shape[0] / dfNew.shape[0] 
    print('percent missing for column {}: {}'.format( col,percent_missing))
```
